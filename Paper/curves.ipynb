```python
import numpy as np
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(42)

# Generate the x-axis values
t = np.linspace(0, 10, 500)

# Generate random processes for the curves
x_1 = np.sin(t) + np.random.normal(0, 0.2, size=t.shape) + 2  # Extreme positive shift
x_2 = np.sin(t) + np.random.normal(0, 0.2, size=t.shape)      # Base curve
x_3 = x_2 + + np.cos(t) + np.random.normal(0, 0.05, size=t.shape)           # Overlaps with x_2
x_4 = np.sin(t) + np.random.normal(0, 0.2, size=t.shape) - 2  # Extreme negative shift

# Plot the curves
plt.figure(figsize=(10, 6))
plt.plot(t, x_1, label="x_1 (extreme positive)", color="blue")
plt.plot(t, x_2, label="x_2 (overlapping)", color="green")
plt.plot(t, x_3, label="x_3 (overlapping)", color="orange")
plt.plot(t, x_4, label="x_4 (extreme negative)", color="red")

# Add labels, legend, and title
plt.xlabel("Time (t)")
plt.ylabel("Value")
plt.title("Generated Curves from Random Processes")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
